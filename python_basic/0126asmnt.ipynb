```python
import glob
import re
from time import sleep


def countContents():
    # read files - whole 1880 ~ 2018
    # path
    name_path  = './name/*'
    file_list = glob.glob(name_path)
    name_list = [item for item in file_list if item.endswith('.txt')]
    year_list = [int(re.sub(r'[^0-9]', '', y)) for y in name_list] # only years

    # file content read
    gender_content = []
    name_content = []

    for i in range(len(name_list)):
        gcount = dict()
        # ncount = dict()
        bcount = 0
        male_dict = dict()
        female_dict = dict()

        with open(name_list[i], 'r') as f:
            data = f.readlines()
            for d in data:
                if d[-1] == '\n':
                    d = d[:-1]
                    name = d.split(',')[0]
                    gender = d.split(',')[1]
                    birth = d.split(',')[2]
                    bcount += int(birth) # 연도별 탄생수의 합산

                    # 성별 별
                    if gender not in gcount:
                        gcount[gender] = 0
                    gcount[gender] += int(birth)

                    # 이름 별
                    if gender == 'F':
                        female_dict[name] = int(birth)
                    else:
                        male_dict[name] = int(birth)
                    # if name not in ncount:
                    #     if gender == 'F':
                    #         male_list.append((name, ncount[name]))
        male_dict = sorted(male_dict.items(), key=lambda x : x[1], reverse=True)
        female_dict = sorted(female_dict.items(), key=lambda x : x[1], reverse=True)

        gender_content.append((year_list[i], gcount))
        name_content.append((year_list[i], male_dict, female_dict))

    return gender_content, name_content

gender_result, name_result = countContents()


# assignment functions
def asgmnt1(gender_result):
    print('과제 1 : name 폴더에 있는 파일들을 활용하여 Mary,F,1980 이런 식의 형태에서 연도별, 성별 별 출생아 수 통계처리 정보를 출력하시오.\n\n')
    for year, gcount in gender_result:
        print(f'연도 : {year}년 - 여성은 {gcount["F"]}명이고 남성은 {gcount["M"]}명 이며, 총 : {sum(gcount.values())}명이고 {"여성이 남성보다" if gcount["F"] > gcount["M"] else "남성이 여성보다"} {abs(gcount["F"]-gcount["M"])}명 많다.')


def asgmnt2(name_result):
    print('과제 2 : name 폴더에 있는 파일들을 활용하여 연도별 인기있는 상위 10개 성별 과 출생아의 이름을 기준으로 통계처리 정보 출력하시오.\n\n')
    # write in csv
    for year, male_dict, female_dict in name_result:
        with open(f'./result/{year}_rank.csv','w') as file:
            # male
            for key, val in male_dict:
                data = f'Male,{key},{val}\n' # '%s,%s\n'  %(year, birth)
                file.write(data)
            # female
            for key, val in female_dict:
                data = f'Female,{key},{val}\n' # '%s,%s\n'  %(year, birth)
                file.write(data)
    print(f'예시 출력 : {name_result[-1][0]}년\n')
    print(f'{name_result[-1][0]}년 상위 10명의 이름과 수')
    for i in range(10):
        print(f'TOP {i+1}\n남자 이름 - {name_result[-1][1][i][0]}, 총 {name_result[-1][1][i][1]}명 / 여자 이름 - {name_result[-1][2][i][0]}, 총 {name_result[-1][2][i][1]}명\n')
    # print(name_result[-1][1][:10]) # male


# function calls
asgmnt1(gender_result)
sleep(1)
print('\n\n')
asgmnt2(name_result)

```
